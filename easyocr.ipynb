```python
import easyocr
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

## Read in images or video

```python
Image1 = 'hello.jpg'
```

```python
reader = easyocr.Reader({'en'}, gpu = False)
result = reader.readtext(Image1)
result
```

 ## Draw Results

```python
top_left = tuple(result[0][0][0])
bottom_right = tuple(result[0][0][2])
text = result[0][1]
font = cv2.FONT_HERSHEY_SIMPLEX
```

```python
img = cv2.imread(Image1)
img = cv2.rectangle(img,top_left,bottom_right,(0,255,0),5)
img = cv2.putText(img,text,top_left, font,2,(152,140,160),3)
plt.imshow(img)
plt.show()
```

## Detecting multiple lines of text

```python
Image2 = 'sign_board.jpg'
reader = easyocr.Reader({'en'}, gpu = False)
result = reader.readtext(Image2)
result
```

```python
img = cv2.imread(Image2)
for detection in result:
    top_left = tuple([int(val) for val in detection[0][0]])
    bottom_right = tuple([int(val) for val in detection[0][2]])
    text = detection[1]
    font = cv2.FONT_HERSHEY_SIMPLEX
    img = cv2.rectangle(img,top_left,bottom_right,(0,255,0),5)
    img = cv2.putText(img,text,top_left,font,2,(255,255,255),3)
    
plt.figure(figsize=(10,10))
plt.imshow(img)
plt.show()
```
